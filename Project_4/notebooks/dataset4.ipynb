- `loc()` & `iloc()` , set values to rows, columns and cell or cells

```python
import pandas as pd
import math
sci = pd.read_csv("../data/salaries-distribution.csv")

sci.loc[4:7, ['industry', 'experience_years', 'salary_usd']]

sci.iloc[4:9, 0:4]

sci.value_counts('industry')

# sci.loc[0, :] = math.nan                to set a row as NaN
sci.loc[75, ['location']] = 'Indian'        #to set a value in a cell

# condition = sci['experience_years'] > 15
# sci.loc[condition, ['industry']] = 'Yash'         this sets the value conditionally

sci['new_col'] = sci['salary_usd'] / 100

sci.drop('new_col', axis=1, inplace=True)
sci

sci.sample(frac=0.25)                # this gives a 30% sample random copy of the df
sci.sample(frac=2, replace=True, ignore_index= True)           # this gives a double the size of df sample random copy ignoring the index

sci['industry']

#  this is new
```
