```python
#lists
lst = [100, 200, 300, 400, 500, 600, 700, 800, 900]     #to declare a list
lst2 = [
    {
        "id": 1,
        "name": "Yash",
        "age": 23,
    }
]

print(lst[:])               #slicing when no start or end is provided outputs the full list
print(lst[2:])              #slicing when start is provided but no end goes from start to the total end of the list
print(lst[:8])              #slicing when end is provided but no start goes from total start to the end of the list
print(lst[-3])              #negative index is to get list value from the end
#print(lst)
# for x in lst:
    #print(x)

# for i, val in enumerate(lst):  print(i, val)
# lst.sort(reverse: True)

print(min(lst))
print(max(lst))
print(sum(lst))
print(lst2[0])
print(lst2[0]['name'])
```

```python
#dictionary
my_dict = {'key1': 'mango', 'key2': 'banana', 'key3': 'kiwi', 'key4': 'apple', 'key5': 'orange'}

my_dict.keys()      #to get keys of a dictionary
my_dict.values()      #to get values of a dictionary
my_dict.items()      #to get items of a dictionary

my_dict['key6'] = 'pomogranate'      #to add keys of a dictionary
my_dict['key6'] = 'pinneaple'       #to update the dictionary

new_dict = my_dict.copy()           #to copy the dictionary

my_dict.pop('key6')                 #to remove the elements
del my_dict['key5']                 #to remove the elements
#my_dict.clear()                    #to clear the dictionary

if "key6" in my_dict: print("exists")  
else: print('doesnotexists')                  #to check if the key exists in the dictionary

for key in new_dict:
    print(key, new_dict[key])                #to loop in the dictionary
print(my_dict, "length", len(my_dict))
print(new_dict, "length", len(new_dict))

nested_dic = {
    "user": {
        "name": "Yash",
        "age": 21
    }
}
print(nested_dic['user']["name"])                   #nested dictionaries

com_dict = {x: x*x for x in range(5)}               #comprehension in dictionaries
print(com_dict)

d1 = {"a" : 1}
d2 = {"b" : 2}

d = d1 | d2                                         #Merge dictionaries
print(d)
```
