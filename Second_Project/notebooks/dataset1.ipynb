```python
import numpy as np
import pandas as pd
math_column = np.random.randint(1, 95, size= 145)
science_column = np.random.randint(1, 97, size= 145)
english_column = np.random.randint(1, 85, size= 145)
social_column = np.random.randint(1, 80, size= 145)
spanish_column = np.random.randint(1, 98, size= 145)
# students_column = 

df_marks = pd.DataFrame({'Math': math_column, 'Science': science_column, 'English': english_column, 'Social': social_column, 'Spanish': spanish_column})
df_marks

```

### Table operations (Math)

```python
#In Coulumns
sum = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].sum()
count = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].count()
mean = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].mean()
stdDeviation = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].std()
max = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].max()
min = df_marks[['Math', 'Science', 'Social', 'English', 'Spanish']].min()

dict = {'sum': sum, 'count': count, 'Mean': mean, 'Standard Deviation': stdDeviation, 'Max': max, 'Min': min}
dict
```

```python
# In rows
df_marks['Total'] = df_marks['Math'] + df_marks['Science'] + df_marks['English'] + df_marks['Social'] + df_marks['Spanish']
df_marks['Average'] = (df_marks['Math'] + df_marks['Science'] + df_marks['English'] + df_marks['Social'] + df_marks['Spanish'])/5
df_marks['Percentage'] = (df_marks['Total'] / 500) * 100
df_marks
```

Sort a Dataframe by column and index

```python
df_marks.sort_values('Math')    #this sorts in descending and by math column
df_marks.sort_values(['Math', 'Science'], ascending=True)    #this sorts in ascending and by first math & science column
```

```python
new_index = np.arange(1, 146)
len(new_index)

df_marks['index'] = new_index                       #a new index column is added at the end of the dataframe
df_marks

df_marks = df_marks.set_index('index')              # the last column, index, is set as index of the dataframe. we can use the inplace=true to mutate the original df.
df_marks

df_marks.sort_index(ascending=False)
```
