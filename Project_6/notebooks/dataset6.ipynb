```python
import pandas as pd
import numpy as np

df_cars = pd.read_csv('../data/Car_sales.csv')


df_cars = df_cars[['Latest_Launch', 'Manufacturer', 'Model', 'Vehicle_type', 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity', 'Fuel_efficiency']]

df_cars.dropna()

# df_cars.groupby(by='Manufacturer').groups

# df_cars.groupby(by='Manufacturer').agg('sum')[['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity', 'Fuel_efficiency']]

# df_cars.groupby(by='Vehicle_type').agg('sum')
# df_cars.groupby(by='Vehicle_type').agg('sum')

df_cars['Latest_Launch'] = df_cars['Latest_Launch'].astype('datetime64[ns]')
years = df_cars['Latest_Launch'].dt.year

df_cars.insert(loc=1, column='Years', value=years)

df_years_mean = df_cars.groupby(by = 'Years', as_index=False)[['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity', 'Fuel_efficiency']].agg('mean')
df_vehicle_sum = df_cars.groupby(by = 'Vehicle_type')[['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity', 'Fuel_efficiency']].agg('sum')

df_vehicle_sum

# df_vehicle_plot = df_vehicle_sum.plot(kind='bar', ylabel='Stats', xlabel= 'Vehicle Type')

df_manufacturer_min_max = df_cars.groupby(by = 'Manufacturer')[['Horsepower', 'Engine_size', 'Fuel_capacity', 'Fuel_efficiency']].agg(['max', 'min'])

df_mmm_plot = df_manufacturer_min_max.head(15).plot(kind = 'bar')
```
